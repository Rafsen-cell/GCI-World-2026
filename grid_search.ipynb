```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.svm import SVC
import numpy as np
```

```python
# load the breast cancer data
cancer = load_breast_cancer()

# Split into training and test data
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=0)

# Prepare parameters for GridSearchSV
param_grid ={
    'C': np.logspace(-3, 2, num=6),
    'gamma': np.logspace(-3, 2, num=6)
}

# Initialize GridSearchSV
gs = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    cv=5
)

# perform hyperparameter tuning and build the best model
gs.fit(X_train, y_train)

# Display result
print('Best cross-validation scores: {:.3f}'.format(gs.best_score_))
print('Best parameters: {}'.format(gs.best_params_))
print('Test set scores: {:.3f}'.format(gs.score(X_test, y_test)))
```
