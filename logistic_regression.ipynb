```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('house_prices.csv')
pd.set_option('display.max_columns', None)
```

```python
columns_to_drop = ['id', 'date', 'waterfront', 'view', ]
```

```python
df = df.drop(columns=columns_to_drop)
```

```python
print(df['condition'].value_counts())
```

```python
# i want to predict condition good or not with other variables
X = df.drop(columns='condition')
y = df['condition']

#split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)

#model
model = LogisticRegression()
model.fit(X_train, y_train)

#view score
print('Training accuracy scores: {:.3f}'.format(model.score(X_train, y_train)))
print('Test accuracy scores: {:.3f}'.format(model.score(X_test, y_test)))
```
