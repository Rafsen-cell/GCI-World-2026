```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# Load breast cancer dataset
cancer = load_breast_cancer()

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, train_size=0.8, random_state=0)

# Call Logistic Regression model
lg = LogisticRegression(random_state=0)
lg.fit(X_train, y_train)

# Make confusion matrix
y_pred = lg.predict(X_test)
m = confusion_matrix(y_test, y_pred)
print(f'Confusion matrix:\n{m}')

# Display score of precission, recall, and f1-score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Display the scores
print(f'Precision:{precision:.3f}')
print(f'Recall:{recall:.3f}')
print(f'F1 Score:{f1:.3f}')
```
