```python
def pareto_analysis(df, sort_by, n):
    df_order = df.sort_values(by=sort_by, ascending=False).copy()
    ranks = df_order[sort_by].rank(method='first', ascending=False)
    df_order['group'] = pd.qcut(ranks, q=n)
    total = df_order[sort_by].sum()
    proportions = df_order.groupby('group')[sort_by].sum()/total
    proportions = proportions.sort_values(ascending=False)
    return proportions
```

```python
pareto_analysis(df=df, sort_by='Sales', n=10)
```
