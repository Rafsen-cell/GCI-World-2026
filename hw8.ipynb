```python
def homework(anime_data_extracted):
    df = anime_data_extracted.copy()

    # Create 10 equal-frequency groups (deciles) based on Members
    df['Decile'] = pd.qcut(
        df['Members'],
        q=10,
        labels=['01', '02', '03', '04', '05', '06', '07', '08', '09', '10']
    )

    # Aggregate MeanScore and TotalMembers per decile
    result = df.groupby('Decile').agg(
        MeanScore=('Score', 'mean'),
        TotalMembers=('Members', 'sum')
    )

    # Cumulative percentage of TotalMembers relative to overall total
    result['CumulativePct'] = (
        result['TotalMembers'].cumsum() / result['TotalMembers'].sum() * 100
    )

    result.index.name = 'Members'

    return result[['MeanScore', 'TotalMembers', 'CumulativePct']]
```
